### Webscraping from website to get the data using selenium

April Dataset

```python
import time
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Define the path of chromedriver with the path to your chromedriver executable
webdriver_service = Service('/Users/basuc/OneDrive/Documents/chromedriver')

# Set Chrome options
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")  # Optional: Disable GPU acceleration
chrome_options.add_argument("--no-sandbox")  # Optional: Add sandbox argument if needed

# Specify the download directory and desired file name
download_directory = '/Users/basuc/Downloads'
file_name = 'ATMAICSAPRIL2022DCE60947F453408BB0B9B12FC25B4FB3'

prefs = {
    "download.default_directory": download_directory,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
}

chrome_options.add_experimental_option("prefs", prefs)

# Create a new instance of the Chrome driver with the configured options
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

try:
    # define the URL of the website
    driver.get('https://www.rbi.org.in/Scripts/ATMView.aspx')

    # Define the download link from the website
    download_link = 'https://rbidocs.rbi.org.in/rdocs/ATM/DOCs/ATMAICSAPRIL2022DCE60947F453408BB0B9B12FC25B4FB3.XLSX'

    # Click on the download link
    driver.get(download_link)

    # Wait for the file to be fully downloaded
    timeout = 60  # Increase the timeout value if needed
    downloaded_file_path = os.path.join(download_directory, file_name)
    while timeout > 0:
        time.sleep(1)
        if os.path.exists(downloaded_file_path):
            break
        timeout -= 1
    else:
        print("Timeout occurred while waiting for the file to download.")

    # Rename the downloaded file
    if os.path.exists(downloaded_file_path):
        new_file_path = os.path.join(download_directory, 'downloaded_file.xlsx')
        os.rename(downloaded_file_path, new_file_path)
        print("File downloaded successfully.")
    else:
        print("File download Successful.")

finally:
    # Quit the driver
    driver.quit()
```
